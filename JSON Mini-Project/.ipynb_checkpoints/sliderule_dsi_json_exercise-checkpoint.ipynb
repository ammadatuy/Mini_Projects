## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,

****
## Answer to Question #1
Find the 10 countries with most projects

```python
# load as Pandas dataframe
json_df = pd.read_json('data/world_bank_projects.json')

# inspect columns to get relevant columns
json_df.columns

# inspect subset of relevant columns
subset_json_df1 = json_df[['countryshortname', 'project_name']]
subset_json_df1.head()

# drop any duplicates
subset_json_df1.drop_duplicates()

# get frequency count of 10 countries with the most projects, grouped by countryshortname 
grouped = subset_json_df1.groupby(['countryshortname'])[['project_name']].count()
grouped.columns = ['project_count']

# find the top 10 countries with most projects and list them in descending oder 
countries_with_most_projects = grouped.sort_values('project_count', ascending=False)[:10]
print(countries_with_most_projects)
```

```python
# plot a bar graph of the 10 countries with the most projects
import matplotlib.pyplot as plt
plot_df = subset_json_df1['countryshortname'].value_counts()[:10]
plot_df.plot(kind='bar')
plt.xlabel('Countries')
plt.ylabel('Frequency')
plt.title('Countries with the Most Projects')
```

## Answer to Question #2
Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# define json string 
json_str = json.load((open('data/world_bank_projects.json')))

# use normalization to create a table from nested mjtheme_namecode column
json_normalized_df1 = json_normalize(json_str, 'mjtheme_namecode')

# create new column 'count' with frequency of project names and group by the code column 
json_normalized_df1['count'] = json_normalized_df1.groupby('code')['code'].transform('count')

# drop any duplicates 
json_normalized_df1 = json_normalized_df1.drop_duplicates()

# get subset of dataframe that does not have empty string 
json_normalized_df1 = json_normalized_df1[json_normalized_df1.name != '']

# set code column as the index 
json_normalized_df1 = json_normalized_df1.set_index('code')

# sort list by descending count frequency, top 10 themes
json_normalized_df1 = json_normalized_df1.sort_values(['count'], ascending=False)[:10]
json_normalized_df1
```

```python
# plot a bar graph of the top 10 project themes
plot_df2 = json_normalized_df1
plot_df2.plot('name', 'count', kind='bar')
plt.xlabel('Project Themes')
plt.ylabel('Frequency')
plt.title('Top 10 Project Themes')
```

## Answer to Question #3
In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import numpy as np

# use normalization to create a table from nested mjtheme_namecode column
json_normalized_df2 = json_normalize(json_str, 'mjtheme_namecode')

# create dictionary with key:code and value:project name 
df_key = {'1': 'Economic management',
          '2': 'Public sector governance',
          '3': 'Rule of law	',
          '4': 'Financial and private sector development',
          '5': 'Trade and integration',
          '6': 'Social protection and risk management',
          '7': 'Social dev/gender/inclusion',
          '8': 'Human development',
          '9': 'Urban development',
          '10': 'Rural development',
          '11': 'Environment and natural resources management'
}

# loop through dictionary df_key and set the correct values in the dataframe 
for code in df_key:
    json_normalized_df2.loc[json_normalized_df2.code == code, ['name']] = df_key[code]

json_normalized_df2
```
